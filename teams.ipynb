```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv('./teams_2020.csv')
```

```python
data
```

```python
weeks = data['Week']
```

```python
weeks
```

```python
scores = data.loc[:, data.columns != 'Week']
```

```python
scores
```

```python
points_for = pd.DataFrame({'team': scores.columns, 'points_for': scores.sum()})
```

```python
points_for
```

```python
wide_form = data.melt('Week', var_name='team', value_name='points_for')
```

```python
(
    alt.Chart(wide_form)
        .mark_line()
        .encode(x='Week:O', y='points_for:Q', color='team:N')
        .properties(width=800, height=300, title='Weekly Scores')
)
```

```python
standings = [
    'Kristen McCaffrey',
    'Killer Clowns',
    'Overconfidentii Vulgaris',
    'In Brees We Trust',
    'Brian\'s Bulldogs',
    'The Sub-Marinos',
    'Spaceballs The Fantasy Team',
    'Birmingham Iron',
    'Team Teem',
    'The Ol Dirty Sock',
    'Green Bay Popik',
    'Paper Lions',
    'Pigskin Syndicate',
    'Untar Stars',
]
```

```python
(alt.Chart(wide_form)
    .mark_boxplot()
    .encode(
        alt.X('team:N', sort=standings), 
        alt.Y('points_for:Q')
    )
    .properties(width=700, height=400, title='Score Distribution by Team (Full Season, Ordered by Standings)'))
```

```python
(alt.Chart(wide_form)
    .mark_bar(tooltip=True)
    .encode(
        alt.X('team:N', sort=standings),
        alt.Y('sum(points_for):Q')
    )
    .properties(title='Total Points For (Ordered by Standings)')
)
```

```python
points_for_stats = pd.DataFrame({'team': scores.columns, 'pf_mean': scores.mean(), 'pf_var': scores.var(), 'pf_std': scores.std(), 'standings': [standings.index(t) + 1 for t in scores.columns]})
```

```python
points_for_stats
```

```python
(alt.Chart(points_for_stats)
    .mark_point(tooltip=True)
    .encode(
        alt.X('pf_mean:Q', scale=alt.Scale(zero=False)), 
        alt.Y('pf_std:Q', scale=alt.Scale(zero=True)), 
        alt.Color('standings:O'), 
        alt.Text('team')
    )
    .properties(
        title='Mean vs. Standard Deviation, Points For'
    )
)
```

```python
draft_order_2020 = [
    'Spaceballs The Fantasy Team',
    'The Ol Dirty Sock',
    'Green Bay Popik',
    'Untar Stars',
    'Team Teem',
    'Killer Clowns',
    'Kristen McCaffrey',
    'Brian\'s Bulldogs',
    'Paper Lions',
    'The Sub-Marinos',
    'Pigskin Syndicate',
    'In Brees We Trust',
    'Birmingham Iron',
    'Overconfidentii Vulgaris',
]
```

```python
standings_2020 = standings
```

```python
draft_order_2019 = [
    'Paper Lions',
    'Green Bay Popik',
    'Stressful Scrumbags',
    'In Brees We Trust',
    'TOM FORKING BRADY',
    'Pigskin Syndicate',
    'The Sub-Marinos',
    'Brian\'s Bulldogs',
    'Killer Clowns',
    'die Waschbar',
    '99RedBalloons',
    'Ray Jaroop',
    'Sony in New England',
    'Team Tranquilo',
]
```

```python
standings_2019 = [
    'Paper Lions',
    'Green Bay Popik',
    'Sony in New England',
    'In Brees We Trust',
    'The Sub-Marinos',
    'TOM FORKING BRADY',
    'Killer Clowns',
    'Ray Jaroop',
    'Brian\'s Bulldogs',
    'Pigskin Syndicate',
    '99RedBalloons',
    'Stressful Scrumbags',
    'die Waschbar',
    'Team Tranquilo',
]
```

```python
draft_vs_standings_data = []
for team in draft_order_2020:
    draft_pos = draft_order_2020.index(team)
    final_pos = standings_2020.index(team)
    draft_vs_standings_data.append([team, draft_pos + 1, final_pos + 1, '2020'])
    
for team in draft_order_2019:
    draft_pos = draft_order_2019.index(team)
    final_pos = standings_2019.index(team)
    draft_vs_standings_data.append([team, draft_pos + 1, final_pos + 1, '2019'])

draft_vs_standings = pd.DataFrame(draft_vs_standings_data, columns=['Team', 'Draft Position', 'Final Standing', 'Year'])
draft_vs_standings
```

```python
(alt.Chart(draft_vs_standings)
    .mark_point(tooltip=True)
    .encode(
        alt.X('Draft Position:Q', scale=alt.Scale(zero=False)), 
        alt.Y('Final Standing:Q', scale=alt.Scale(zero=True)), 
        alt.Color('Year:O')
    )
    .properties(
        title='Draft Position vs. Final Standings, 2019 and 2020'
    )
)
```
