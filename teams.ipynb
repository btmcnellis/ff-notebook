```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv('./teams_2020.csv')
```

```python
data
```

```python
weeks = data['Week']
```

```python
weeks
```

```python
scores = data.loc[:, data.columns != 'Week']
```

```python
scores
```

```python
points_for = pd.DataFrame({'team': scores.columns, 'points_for': scores.sum()})
```

```python
points_for
```

```python
wide_form = data.melt('Week', var_name='team', value_name='points_for')
```

```python
(
    alt.Chart(wide_form)
        .mark_line()
        .encode(x='Week:O', y='points_for:Q', color='team:N')
        .properties(width=800, height=300, title='Weekly Scores')
)
```

```python
standings = [
    'Killer Clowns',
    'Brian\'s Bulldogs',
    'Kristen McCaffrey',
    'Birmingham Iron',
    'Spaceballs The Fantasy Team',
    'Paper Lions',
    'Overconfidentii Vulgaris',
    'The Sub-Marinos',
    'In Brees We Trust',
    'Green Bay Popik',
    'Pigskin Syndicate',
    'Untar Stars',
    'The Ol Dirty Sock',
    'Team Teem'
]
```

```python
(alt.Chart(wide_form)
    .mark_boxplot()
    .encode(
        alt.X('team:N', sort=standings), 
        alt.Y('points_for:Q')
    )
    .properties(width=700, height=400, title='Score Distribution by Team (Through Week 6, Ordered by Standings)'))
```

```python
(alt.Chart(wide_form)
    .mark_bar()
    .encode(
        alt.X('team:N', sort=standings),
        alt.Y('sum(points_for):Q')
    )
    .properties(title='Total Points For (Through Week 6, Ordered by Standings)')
)
```
